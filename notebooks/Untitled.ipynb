```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation,rc
from IPython.display import HTML

plt.rcParams['animation.ffmpeg_path']=r'C:\Users\jayke\jupyter_notebooks\ffmpeg-20190926-87ddf9f-win64-static\bin\ffmpeg.exe'
rc('animation', html='html5')
```

```python
filename=r'C:\Users\jayke\source\repos\waves\out\build\x64-Debug\waves\out.txt'

rod_data=[]
with open(filename) as f:
    for l in f:
        temp=[]
        for el in l.split():
            temp.append(float(el))

        rod_data.append(temp)
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(0, 100), ylim=(-1.5, 1.5))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    line.set_data(range(0,len(rod_data[i])), rod_data[i])
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=200, interval=20)
```
